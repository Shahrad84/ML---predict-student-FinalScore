```python
data = source_df.copy()
df_label = source_df["FinalScore"].copy()
data = data.drop(["ExtracurricularActivities", "Student", "FinalScore"], axis = 1)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(data, df_label)
```

```python
from sklearn.metrics import mean_squared_error
predicted = lin_reg.predict(data)
lin_rmse = mean_squared_error(df_label, predicted)
lin_rmse = np.sqrt(lin_rmse)
lin_rmse
```
